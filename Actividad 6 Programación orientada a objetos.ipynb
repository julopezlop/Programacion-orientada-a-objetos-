```python
import tkinter as interfaz
from tkinter import ttk
from tkinter import messagebox
from tkinter import Frame
class CalculadorNomina:
    def __init__(self):
        self.ventana = interfaz.Tk()
        self.ventana.geometry(f'300x500')
        self.ventana.resizable(False, False)
        self.ventana.title("Calculador de nómina")
        interfaz.Label(self.ventana, text="Nómina", font= ("Helvetica", 12, "italic")).grid(row=0, column=1)
        # self.ventana.rowconfigure(0, minsize=20)
        # self.ventana.columnconfigure(0, minsize=20)
        self.opciones = ["Calcular Nómina"]
        self.empleados = []
        self.ventana2 = None
        self.combo = ttk.Combobox(self.ventana, values=self.opciones, state="readonly")
        self.combo.current(0)
        self.combo.grid(column= 1 , pady=20)
        self.combo.bind("<<ComboboxSelected>>", self.abrir_ventana)
        
        interfaz.Label(self.ventana, text="Agregar Empleado", font= ("Helvetica", 12, "italic")).grid(row=2, column=1)
        interfaz.Label(self.ventana, text="Nombre:").grid(row=3, column=0, padx=5, pady=5)
        self.nombre_entrada = interfaz.Entry(self.ventana)
        self.nombre_entrada.grid(row=3, column=1, padx=5, pady=5)
        interfaz.Label(self.ventana, text="Apellido:").grid(row=4, column=0, padx=5, pady=5)
        self.apellido_entrada = interfaz.Entry(self.ventana)
        self.apellido_entrada.grid(row=4, column=1, padx=5, pady=5)
        interfaz.Label(self.ventana, text="Cargo:").grid(row=5, column=0, padx=5, pady=5)
        self.cargo_entrada = interfaz.Entry(self.ventana)
        self.cargo_entrada.grid(row=5, column=1, padx=5, pady=5)
        interfaz.Label(self.ventana, text="Genero:").grid(row=6, column=0, padx=5, pady=5)
        self.opcionesGenero = ["Masculino", "Femenino", "Bolsa D1", "Pinguino Transexual"]
        self.combo_genero = ttk.Combobox(self.ventana, values=self.opcionesGenero, state="readonly")
        self.combo_genero.current(0)
        self.combo_genero.grid(row=6, column=1, padx=5, pady=5)

        interfaz.Label(self.ventana, text="Días trabajados:").grid(row=8, column=0, padx=5, pady=5)
        vcmd = (self.ventana.register(self.validar_dia), "%P")
        self.dias_trabajados_spinner = interfaz.Spinbox(self.ventana, from_=1, to=31, width=5,validate="focusout", validatecommand=vcmd)
        self.dias_trabajados_spinner.grid(row=8, column=1, padx=5, pady=5)
        interfaz.Label(self.ventana, text="Otros ingresos:").grid(row=9, column=0, padx=5, pady=5)
        self.otros_ingresos_entrada = interfaz.Entry(self.ventana)
        self.otros_ingresos_entrada.grid(row=9, column=1, padx=5, pady=5)

        interfaz.Label(self.ventana, text="Pagos salud:").grid(row=10, column=0, padx=5, pady=5)
        self.pagos_salud_entrada = interfaz.Entry(self.ventana)
        self.pagos_salud_entrada.grid(row=10, column=1, padx=5, pady=5)
        interfaz.Label(self.ventana, text="Aporte pensional:").grid(row=11, column=0, padx=5, pady=5)
        self.pensiones_entrada = interfaz.Entry(self.ventana)
        self.pensiones_entrada.grid(row=11, column=1, padx=5, pady=5)
        boton_guardar = interfaz.Button(self.ventana, text="Guardar empleado",command=self.guardar_empleado)
        boton_guardar.grid(row=12, column=1, pady=15)
        self.ventana.mainloop()

    def validar_dia(self, nuevo_valor):
        """Valida que el valor sea un número entre 1 y 31"""
        if nuevo_valor == "":  
            return True  

        try:
            numero = int(nuevo_valor)
            if 1 <= numero <= 31:
                return True
            else:
                raise ValueError
        except:
            messagebox.showerror("Error", "El día debe ser un número entre 1 y 31.")
            self.dias_trabajados_spinner.delete(0, "end")
            return False
    def guardar_empleado(self):
        nombre = self.nombre_entrada.get().strip()
        apellido = self.apellido_entrada.get().strip()
        cargo = self.cargo_entrada.get().strip()
        genero = self.combo_genero.get().strip()
        # salario = self.salarioDia_entrada.get().strip()
        dias = self.dias_trabajados_spinner.get().strip()
        otros_ing = self.otros_ingresos_entrada.get().strip()
        salud = self.pagos_salud_entrada.get().strip()
        pension = self.pensiones_entrada.get().strip()
        self.nombre_entrada.delete(0,interfaz.END)
        self.apellido_entrada.delete(0,interfaz.END)
        self.cargo_entrada.delete(0,interfaz.END)
        self.pagos_salud_entrada.delete(0,interfaz.END)
        self.pensiones_entrada.delete(0,interfaz.END)
        self.otros_ingresos_entrada.delete(0,interfaz.END)
        
        
        # Validación
        if not (nombre and apellido and cargo and dias and otros_ing and salud and pension):
            messagebox.showerror("Error", "Todos los campos son obligatorios.")
            return

        # Almacenar empleado
        empleado = (nombre, apellido, cargo, genero, dias, otros_ing, salud, pension)
        self.empleados.append(empleado)

        messagebox.showinfo("Guardado", "Empleado registrado correctamente.")

        # Si ventana2 está abierta, actualiza la tabla
        if self.ventana2 is not None:
            self.actualizar_tabla()

    def validar_numero(self, texto):
        if texto == "":
            return False
        texto = texto.replace(",", ".")
        try:
            float(texto)
            return True
        except ValueError:
            return False
        
    def abrir_ventana(self, event):
        if self.ventana2 is not None:
            self.ventana2.destroy()

        self.ventana2 = interfaz.Toplevel()
        self.ventana2.title("Empleados registrados")
        self.ventana2.geometry("900x350")

        self.table = ttk.Treeview(self.ventana2)
        self.table.grid(row=0, column=0, padx=10, pady=10)

        columnas = ["Nombre", "Apellidos", "Cargo", "Género","Días trabajados", "Otros ingresos", "Pagos salud", "Aporte pensional"]

        self.table["columns"] = columnas
        self.table.column("#0", width=0, stretch=interfaz.NO)

        for col in columnas:
            self.table.column(col, anchor=interfaz.W, width=110)
            self.table.heading(col, text=col, anchor=interfaz.W)

        # Mostrar empleados guardados
        self.actualizar_tabla()

        interfaz.Button(self.ventana2, text="Cerrar", command=self.ventana2.destroy).grid(row=1, column=0, pady=10) 
    def actualizar_tabla(self):
        # Limpiar la tabla
        for item in self.table.get_children():
            self.table.delete(item)

        # Insertar empleados
        for emp in self.empleados:
            self.table.insert("", "end", values=emp)

x = CalculadorNomina()
```

```python
import tkinter as tk
from tkinter import ttk, messagebox
from datetime import datetime


class Hotel:
    def __init__(self):
        # Configuración principal
        self.root = tk.Tk()
        self.root.title("Gestión Hotel - 10 Habitaciones")
        self.root.geometry("420x220")
        self.root.resizable(False, False)

        # Inicializar habitaciones (1-10)
        self.habitaciones = {}
        for i in range(1, 11):
            price = 120000 if i <= 5 else 160000
            self.habitaciones[i] = {
                "price": price,
                "ocupada": False,
                "guest": None  
            }

        # Menú
        menubar = tk.Menu(self.root)
        menu_acciones = tk.Menu(menubar, tearoff=0)
        menu_acciones.add_command(label="Consultar habitaciones", command=self.ventana_consultar)
        menu_acciones.add_command(label="Salida de huéspedes", command=self.ventana_salida_pedir_habitacion)
        menubar.add_cascade(label="Acciones", menu=menu_acciones)
        self.root.config(menu=menubar)
        tk.Label(self.root, text="Hotel - 10 habitaciones", font=("Helvetica", 14)).pack(pady=12)
        tk.Label(self.root, text="Use el menú 'Acciones' para gestionar ingresos y salidas.").pack(pady=6)
        self.win_consulta = None

        self.root.mainloop()

    def ventana_consultar(self):
        # Si ya está abierta, traer al frente
        if self.win_consulta is not None and tk.Toplevel.winfo_exists(self.win_consulta):
            self.win_consulta.lift()
            return

        self.win_consulta = tk.Toplevel(self.root)
        self.win_consulta.title("Consultar habitaciones")
        self.win_consulta.geometry("760x350")
        self.win_consulta.resizable(False, False)

        # Treeview para mostrar habitaciones y estado
        columns = ("Nº", "Precio", "Estado", "Nombre", "Apellidos", "Documento", "Fecha ingreso")
        tree = ttk.Treeview(self.win_consulta, columns=columns, show="headings", height=12)
        for col in columns:
            tree.heading(col, text=col)
            # tamaños aproximados
            if col == "Nº":
                tree.column(col, width=40, anchor=tk.CENTER)
            elif col == "Precio":
                tree.column(col, width=90, anchor=tk.E)
            elif col == "Estado":
                tree.column(col, width=100, anchor=tk.CENTER)
            else:
                tree.column(col, width=140, anchor=tk.W)

        tree.grid(row=0, column=0, columnspan=4, padx=10, pady=10)

        # Llenar la tabla
        def llenar_tabla():
            for it in tree.get_children():
                tree.delete(it)
            for num, data in self.habitaciones.items():
                estado = "No disponible" if data["ocupada"] else "Disponible"
                nombre = data["guest"]["nombre"] if data["guest"] else ""
                apellidos = data["guest"]["apellidos"] if data["guest"] else ""
                doc = data["guest"]["doc"] if data["guest"] else ""
                fecha_ing = data["guest"]["fecha_ingreso"].strftime("%d/%m/%Y") if data["guest"] else ""
                tree.insert("", "end", values=(num, f"${data['price']:,}", estado, nombre, apellidos, doc, fecha_ing))

        llenar_tabla()

        # Selección de habitación para ocupar
        tk.Label(self.win_consulta, text="Seleccione nº habitación para ocupar:").grid(row=1, column=0, padx=10, sticky="w")
        spin_hab = tk.Spinbox(self.win_consulta, from_=1, to=10, width=5)
        spin_hab.grid(row=1, column=1, padx=5, sticky="w")

        btn_ocupar = tk.Button(self.win_consulta, text="Ocupar habitación", command=lambda: self.iniciar_ingreso(int(spin_hab.get()), llenar_tabla))
        btn_ocupar.grid(row=1, column=2, padx=5)

        btn_refrescar = tk.Button(self.win_consulta, text="Refrescar", command=llenar_tabla)
        btn_refrescar.grid(row=1, column=3, padx=5)


    def iniciar_ingreso(self, numero, callback_refrescar=None):
        # Validar número
        if numero not in self.habitaciones:
            messagebox.showerror("Error", "Número de habitación inválido.")
            return

        hab = self.habitaciones[numero]
        if hab["ocupada"]:
            messagebox.showerror("Error", f"Habitación {numero} ya está ocupada.")
            return

        # Ventana para ingresar datos del huésped
        win = tk.Toplevel(self.root)
        win.title(f"Ingreso - Habitación {numero} (Precio: ${hab['price']:,})")
        win.geometry("420x300")
        win.resizable(False, False)

        # Labels + entries
        tk.Label(win, text=f"Habitación {numero} - Precio por día: ${hab['price']:,}", font=("Helvetica", 10, "bold")).grid(row=0, column=0, columnspan=2, pady=8)

        tk.Label(win, text="Fecha ingreso (dd/mm/aaaa):").grid(row=1, column=0, sticky="e", padx=5, pady=5)
        entry_fecha = tk.Entry(win)
        entry_fecha.grid(row=1, column=1, padx=5, pady=5)

        tk.Label(win, text="Nombre:").grid(row=2, column=0, sticky="e", padx=5, pady=5)
        entry_nombre = tk.Entry(win)
        entry_nombre.grid(row=2, column=1, padx=5, pady=5)

        tk.Label(win, text="Apellidos:").grid(row=3, column=0, sticky="e", padx=5, pady=5)
        entry_apellidos = tk.Entry(win)
        entry_apellidos.grid(row=3, column=1, padx=5, pady=5)

        tk.Label(win, text="Documento ID:").grid(row=4, column=0, sticky="e", padx=5, pady=5)
        entry_doc = tk.Entry(win)
        entry_doc.grid(row=4, column=1, padx=5, pady=5)

        # Validaciones de formato
        def validar_texto(txt):
            return txt.strip() != "" and all(ch.isalpha() or ch.isspace() or ch == "-" for ch in txt)

        def validar_doc(txt):
            return txt.strip() != "" and txt.strip().isdigit()

        def validar_fecha(texto):
            try:
                dt = datetime.strptime(texto.strip(), "%d/%m/%Y")
                return True, dt
            except Exception:
                return False, None

        def registrar():
            fecha_str = entry_fecha.get().strip()
            nombre = entry_nombre.get().strip()
            apellidos = entry_apellidos.get().strip()
            doc = entry_doc.get().strip()

            campos_vacios = []
            if not fecha_str: campos_vacios.append("Fecha ingreso")
            if not nombre: campos_vacios.append("Nombre")
            if not apellidos: campos_vacios.append("Apellidos")
            if not doc: campos_vacios.append("Documento ID")
            if campos_vacios:
                messagebox.showerror("Error", f"Los siguientes campos son obligatorios: {', '.join(campos_vacios)}")
                return

            # Validaciones
            ok_fecha, fecha_dt = validar_fecha(fecha_str)
            if not ok_fecha:
                messagebox.showerror("Error", "Formato de fecha inválido. Use dd/mm/aaaa.")
                return

            if not validar_texto(nombre):
                messagebox.showerror("Error", "Nombre inválido (solo letras, espacios o guiones).")
                return
            if not validar_texto(apellidos):
                messagebox.showerror("Error", "Apellidos inválidos (solo letras, espacios o guiones).")
                return
            if not validar_doc(doc):
                messagebox.showerror("Error", "Documento inválido (solo números).")
                return

            # Registrar ocupación
            self.habitaciones[numero]["ocupada"] = True
            self.habitaciones[numero]["guest"] = {
                "nombre": nombre,
                "apellidos": apellidos,
                "doc": doc,
                "fecha_ingreso": fecha_dt
            }

            messagebox.showinfo("Ingreso registrado", f"Habitación {numero} ocupada correctamente.")
            win.destroy()
            if callback_refrescar:
                callback_refrescar()

        btn_registrar = tk.Button(win, text="Registrar ingreso", command=registrar)
        btn_registrar.grid(row=5, column=0, columnspan=2, pady=12)

    def ventana_salida_pedir_habitacion(self):
        win = tk.Toplevel(self.root)
        win.title("Salida de huéspedes - Seleccione habitación")
        win.geometry("380x140")
        win.resizable(False, False)

        tk.Label(win, text="Número de habitación a entregar:").grid(row=0, column=0, padx=8, pady=12)
        spin = tk.Spinbox(win, from_=1, to=10, width=5)
        spin.grid(row=0, column=1, padx=5, pady=12)

        def siguiente():
            try:
                num = int(spin.get())
            except Exception:
                messagebox.showerror("Error", "Número de habitación inválido.")
                return

            if num not in self.habitaciones:
                messagebox.showerror("Error", "Número de habitación inválido.")
                return

            if not self.habitaciones[num]["ocupada"]:
                messagebox.showerror("Error", f"Habitación {num} no está ocupada.")
                return

            win.destroy()
            self.ventana_salida_detalle(num)

        tk.Button(win, text="Continuar", command=siguiente).grid(row=1, column=0, columnspan=2, pady=10)

    def ventana_salida_detalle(self, numero):
        data = self.habitaciones[numero]
        guest = data["guest"]
        if not guest:
            messagebox.showerror("Error", "Error interno: huésped no encontrado.")
            return

        win = tk.Toplevel(self.root)
        win.title(f"Salida - Habitación {numero}")
        win.geometry("520x360")
        win.resizable(False, False)

        tk.Label(win, text=f"Habitación {numero} - Precio por día: ${data['price']:,}", font=("Helvetica", 10, "bold")).grid(row=0, column=0, columnspan=2, pady=8)
        tk.Label(win, text=f"Huésped: {guest['nombre']} {guest['apellidos']} (ID: {guest['doc']})").grid(row=1, column=0, columnspan=2, pady=4)
        tk.Label(win, text=f"Fecha ingreso: {guest['fecha_ingreso'].strftime('%d/%m/%Y')}").grid(row=2, column=0, columnspan=2, pady=4)

        # Fecha de salida
        tk.Label(win, text="Fecha salida (dd/mm/aaaa):").grid(row=3, column=0, sticky="e", padx=5, pady=8)
        entry_fecha_salida = tk.Entry(win)
        entry_fecha_salida.grid(row=3, column=1, padx=5, pady=8)

        # Labels calculo
        lbl_dias = tk.Label(win, text="Días alojamiento: -")
        lbl_dias.grid(row=4, column=0, columnspan=2, pady=6)
        lbl_total = tk.Label(win, text="Total a pagar: -")
        lbl_total.grid(row=5, column=0, columnspan=2, pady=6)

        # Botón para calcular
        def calcular():
            fecha_str = entry_fecha_salida.get().strip()
            try:
                fecha_salida = datetime.strptime(fecha_str, "%d/%m/%Y")
            except Exception:
                messagebox.showerror("Error", "Formato de fecha de salida inválido. Use dd/mm/aaaa.")
                return

            fecha_ing = guest["fecha_ingreso"]
            if fecha_salida <= fecha_ing:
                messagebox.showerror("Error", "La fecha de salida debe ser mayor a la fecha de ingreso.")
                return

            dias = (fecha_salida - fecha_ing).days
            # Según consigna: total = días * precio (days = número de noches/estancias)
            total = dias * data["price"]

            lbl_dias.config(text=f"Días alojamiento: {dias}")
            lbl_total.config(text=f"Total a pagar: ${total:,}")

            # Habilitar botón registrar salida
            btn_registrar.config(state=tk.NORMAL)

        btn_calcular = tk.Button(win, text="Calcular total", command=calcular)
        btn_calcular.grid(row=6, column=0, pady=12, padx=5)


        def registrar_salida():
            self.habitaciones[numero]["ocupada"] = False
            self.habitaciones[numero]["guest"] = None
            messagebox.showinfo("Salida registrada", f"Salida registrada. Habitación {numero} queda disponible.")
            win.destroy()
            if self.win_consulta is not None and tk.Toplevel.winfo_exists(self.win_consulta):
               
                try:
                    self.win_consulta.destroy()
                except Exception:
                    pass
                self.win_consulta = None

        btn_registrar = tk.Button(win, text="Registrar salida", state=tk.DISABLED, command=registrar_salida)
        btn_registrar.grid(row=6, column=1, pady=12, padx=5)

if __name__ == "__main__":
    Hotel()
```
