```python
import tkinter as tk
from tkinter import messagebox
import statistics
class Notas:
    def __init__(self,ventana):
        self.lista_notas = []
        self.ventana = ventana
        self.ventana.title = ("Cálculo notas:")
        self.datos_entrada = []


        for i in range(5):
            tk.Label(ventana, text=f"Nota {i+1}:").grid(row=i, column=0, padx= 10, pady = 5)
            dato_entrada = tk.Entry(ventana)
            dato_entrada.grid(row=i, column=1, padx=10, pady= 5)
            self.datos_entrada.append(dato_entrada)

        tk.Button(ventana, text="Calcular", command=self.calcular).grid(row=6, column=1, padx=5, pady=5)
        self.resultado_promedio = tk.Label(ventana, text="Promedio: ")
        self.resultado_promedio.grid(row=7, column=0, columnspan=1)
        self.resultado_desv = tk.Label(ventana, text="Desviación estándar: ")
        self.resultado_desv.grid(row=8, column=0, columnspan=1)
        self.resultado_max = tk.Label(ventana, text="Mayor nota: ")
        self.resultado_max.grid(row=9, column=0, columnspan=1)
        self.resultado_min = tk.Label(ventana, text="Menor nota: ")
        self.resultado_min.grid(row=10, column=0, columnspan=1)

    def calcular(self):
        try:
            self.lista_notas = []
            for dato_entrada in self.datos_entrada:
                self.lista_notas.append(float(dato_entrada.get()))
            if len(self.lista_notas) != 5:
                raise ValueError("Ingresar 5 notas")
            promedio = sum(self.lista_notas)/len(self.lista_notas)
            desviacion = statistics.stdev(self.lista_notas)
            notaMax = max(self.lista_notas)    
            notaMin = min(self.lista_notas)   

            self.resultado_promedio.config(text=f"Promedio: {promedio}") 
            self.resultado_desv.config(text=f"Deviación estándar: {round(desviacion,3)}") 
            self.resultado_max.config(text=f"Mayor nota: {notaMax}") 
            self.resultado_min.config(text=f"Menor nota: {notaMin}") 
        except ValueError:
            messagebox.showerror("Error", "Ingresar valores númericos")    
                
if __name__== "__main__":
    ventana = tk.Tk()
    app = Notas(ventana)
    ventana.mainloop()        
```

```python
import tkinter as tk
from tkinter import messagebox
import math
class Personas:
    def __init__(self,ventana):
        self.ventana = ventana
        self.ventana.title("Cálculo volumen y superficie")
        self.datos_cilindro = []
        self.datos_esfera = []
        self.datos_piramide = []
        
        tk.Label(ventana, text="Cilindro", font=("Helvetica", 12, "italic")).grid(row=0, column=0, pady=(10,10))
        tk.Label(ventana, text="Radio: ").grid(row=1, column=0, padx= 10, pady = 5)
        radioCilindro = tk.Entry(ventana)
        radioCilindro.grid(row=1, column=1, padx=10, pady= 5)
        self.datos_cilindro.append(radioCilindro)
        
        tk.Label(ventana, text="Altura: ").grid(row=2, column=0, padx= 10, pady = 5)
        alturaCilindro = tk.Entry(ventana)
        alturaCilindro.grid(row=2, column=1, padx=10, pady= 5)
        self.datos_cilindro.append(alturaCilindro)

        tk.Button(ventana, text="Calcular", command=self.calculosCilindro).grid(row=4, column=2, padx=5, pady=5)    
        self.resultado_volumenCilindro = tk.Label(ventana, text="Volumen Cilindro: ")
        self.resultado_volumenCilindro.grid(row = 3, column =0, columnspan=1)
        self.resultado_superficieCilindro = tk.Label(ventana, text="Superficie Cilindro: ")
        self.resultado_superficieCilindro.grid(row= 4, column= 0, columnspan=1)

        tk.Label(ventana, text="Esfera", font=("Helvetica", 12, "italic")).grid(row=5, column=0, pady=(10,10))
        tk.Label(ventana, text="Radio: ").grid(row=6, column=0, padx= 10, pady = 5)
        radioEsfera = tk.Entry(ventana)
        radioEsfera.grid(row=6, column=1, padx=10, pady= 5)
        self.datos_esfera.append(radioEsfera)

        tk.Button(ventana, text="Calcular", command=self.calculosEsfera).grid(row=7, column=2, columnspan=1)
        self.resultado_volumenEsfera = tk.Label(ventana, text="Volumen Esfera: ")
        self.resultado_volumenEsfera.grid(row=7, column=0, columnspan=1)

        tk.Label(ventana, text="Piramide", font=("Helvetica", 12, "italic")).grid(row=8, column=0, pady=(10,10))
        tk.Label(ventana, text="Lado Base: ").grid(row=9, column=0, padx= 10, pady = 5)
        ladoBase = tk.Entry(ventana)
        ladoBase.grid(row=9, column=1, padx=10, pady= 5)
        self.datos_piramide.append(ladoBase)

        tk.Label(ventana, text="Altura Piramide: ").grid(row=10, column=0, padx= 10, pady = 5)
        AlturaPiramide = tk.Entry(ventana)
        AlturaPiramide.grid(row=10, column=1, padx=10, pady= 5)
        self.datos_piramide.append(AlturaPiramide)

        tk.Label(ventana, text="Apotema Piramide: ").grid(row=11, column=0, padx= 10, pady = 5)
        apotemaPiramide = tk.Entry(ventana)
        apotemaPiramide.grid(row=11, column=1, padx=10, pady= 5)
        self.datos_piramide.append(apotemaPiramide)

        tk.Button(ventana, text="Calcular", command=self.calculosPiramide).grid(row=13, column=2, columnspan=1)
        self.resultado_volumenPiramide = tk.Label(ventana, text="Volumen Piramide: ")
        self.resultado_volumenPiramide.grid(row=12, column=0, columnspan=1)
        self.resultado_superficiePiramide = tk.Label(ventana, text="Volumen Piramide: ")
        self.resultado_superficiePiramide.grid(row=13, column=0, columnspan=1)

    
    def calculosCilindro(self):
        try:
            r = float(self.datos_cilindro[0].get()) 
            h = float(self.datos_cilindro[1].get()) 
            volumenCilindro = math.pi*r**2*h
            superficieCilindro = 2*math.pi*r*(r+h)
            self.resultado_volumenCilindro.config(text=f"Volumen Cilindro: {round(volumenCilindro, 2)}")
            self.resultado_superficieCilindro.config(text=f"Superficie Cilindro: {round(superficieCilindro, 2)}")

        except ValueError:
            messagebox.showerror("Error", "Ingrese valores numéricos válidos.")

    def calculosEsfera(self):
        try:
            r = float(self.datos_esfera[0].get())
            volumenEsfera = 4/3 *math.pi * r**3
            self.resultado_volumenEsfera.config(text=f"Volumen Esfera: {round(volumenEsfera,2)}")    
        
        except ValueError:
            messagebox.showerror("Error", "Ingrese valores numéricos válidos.")
    def calculosPiramide(self):
        try:
            lb = float(self.datos_piramide[0].get())
            a = float(self.datos_piramide[1].get())
            ap = float(self.datos_piramide[2].get())
            volumenPiramide = 1/3 * lb**2*a
            superficiePiramide = lb**2 + 2*lb*a
            self.resultado_volumenPiramide.config(text=f"Volumen Piramide: {round(volumenPiramide,2)}")    
            self.resultado_superficiePiramide.config(text=f"Superficie Piramide: {round(superficiePiramide,2)}")    
        
        except ValueError:
            messagebox.showerror("Error", "Ingrese valores numéricos válidos.")

if __name__ == "__main__":
    ventana = tk.Tk()
    app = Personas(ventana)
    print()
    ventana.mainloop()   
```
